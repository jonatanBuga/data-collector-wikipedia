```python
import pandas as pd
import requests
import random
import json
from bs4 import BeautifulSoup
from tqdm import tqdm
```

```python
#@title API for page id
from numpy.lib.nanfunctions import nanmedian

# Construct the API URL
def get_api_url(name):
  api_url = f'https://en.wikipedia.org/w/api.php?action=query&titles={name}&prop=pageprops&format=json'

# Make the API request and convert the response to JSON
  response = requests.get(api_url).json()

# # Extract the page ID from the JSON response
  try:
    page_id = next(iter(response['query']['pages'].values()))['pageprops']['wikibase_item']
    print(f"The Wikidata page ID for {name} is {page_id}")
    return page_id
  except :
    return
```

```python
#@title collect data from wikidata
from bs4 import BeautifulSoup
def wikidataInfo(UniqueId,arrtotable):
  Url = f"https://www.wikidata.org/wiki/{UniqueId}"
  html = requests.get(Url).content.decode('utf-8')
  soup = BeautifulSoup(html, 'html.parser')
  l = soup.findAll('a')
  name=soup.findAll('span',{'class':'wikibase-title-label'})

  wikidata=soup.findAll('span',{'class':'wikibase-title-id'})
  nameN=name[0].text
  discrip=soup.findAll("span",{'class':'wikibase-descriptionview-text'})
  description=discrip[0].text

  entries_div=soup.findAll("div",{"class":"wikibase-sitelinkgroupview"})
  entries=entries_div[0].findAll("li",{"class":"wikibase-sitelinkview"})
  for i in entries:
    lang=i.findAll("span",{'class':'wikibase-sitelinkview-siteid'})
    wiki=lang[0].text.index('wiki')
    lang_shortcut=lang[0].text[:wiki]


    nameinlang=i.findAll('a')
    result=(nameN,wikidata[0].text,description,lang_shortcut,lang[0].get('title'),nameinlang[0].text)
    arrtotable.append(result)
```

```python
#@title Create Pdtable
from google.colab import files
def createTable(names,namepage):
  Data_name=[]
  for name in names:
    name_to_check = get_api_url(name)
    if name_to_check:
        print('test')
        wikidataInfo(name_to_check,Data_name)
    else:
      continue
  results_df = pd.DataFrame(Data_name, columns=['Label', 'WikiData ID', 'English Description', 'Language','Wiki ShortLang','Entry'])
  if namepage =='Given names':
    results_df.to_csv('Given names.csv', index=False, encoding='utf-8-sig')
    files.download('Given names.csv')
  else:
    results_df.to_csv('Surnames.csv', index=False, encoding='utf-8-sig')
    files.download('Surnames.csv')
```

```python
#@title initilazie pages and names
import requests
from bs4 import BeautifulSoup
def init_page(namepage):
  url = f"https://en.wikipedia.org/wiki/Category:{namepage}"
  list_of_names=[]
  while True:
    response = requests.get(url)
    soup = BeautifulSoup(response.text,"html.parser")
    tags_of_names = soup.find("div",id="mw-pages").find("div",{"class":"mw-category mw-category-columns"}).findAll('a')
    for j in tags_of_names:
          list_of_names.append(j.text)
    next_page = soup.find('div',id='mw-pages').find('a',string='next page')
    if next_page is None:
      break;
    url = 'https://en.wikipedia.org'+next_page['href']
  list_of_names = list_of_names[4:]
  print('finish init')
  return list_of_names

```

```python
first='Given names'
Second = 'Surnames'
li_of_names=init_page(Second)
createTable(li_of_names,Second)#for running the script we need to slicing li_of_names array [] for example: li_of_names[:20000]
```
